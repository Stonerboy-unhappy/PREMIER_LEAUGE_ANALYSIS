```python
i = j = 1
# Create an empty subplot:
fig = make_subplots(rows=1, cols=1,
                    shared_xaxes=False,
                    vertical_spacing=0.1)
# Add bar plot for goals conceded in all subplots:
for col in num_cols:
    fig.add_trace(go.Bar(x=total_df['Name'],
                        y=total_df[col],
                        text=total_df[col],
                        textposition="inside",
                        name=col),
                row=i, col=j)
    j += 1
    if j > 1:
        j = 1
        i += 1
    if i > 1:
        i = 1
fig.update_layout(width=1500,height=800, title='Premier Leauge Top 10 Goal Scorer Goals')
fig.show()
```
